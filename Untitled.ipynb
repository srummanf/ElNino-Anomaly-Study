```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = [
    "Entrepreneurship", "Consulting", "Coaching", "Marketing",
    "Technology", "Education", "Real Estate", "Legal", "Others"
]
percentages = [28, 14, 9, 8, 7, 6, 5, 4, 19]

# Function to show and save plots
def show_and_save_plot(fig, ax, title):
    ax.set_title(title)
    plt.tight_layout()
    plt.show()

# 1. Pie Chart
fig, ax = plt.subplots()
ax.pie(percentages, labels=categories, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
show_and_save_plot(fig, ax, "Pie Chart")

# 2. Bar Chart
fig, ax = plt.subplots()
ax.bar(categories, percentages, color=plt.cm.tab10.colors)
ax.set_ylabel("Percentage")
ax.set_title("Bar Chart")
plt.xticks(rotation=45)
plt.show()

# 3. Horizontal Bar Chart
fig, ax = plt.subplots()
ax.barh(categories, percentages, color=plt.cm.tab10.colors)
ax.set_xlabel("Percentage")
ax.set_title("Horizontal Bar Chart")
plt.show()

# 4. Line Chart
fig, ax = plt.subplots()
ax.plot(categories, percentages, marker='o', color='blue')
ax.set_ylabel("Percentage")
ax.set_title("Line Chart")
plt.xticks(rotation=45)
plt.show()

# 5. Scatter Plot
fig, ax = plt.subplots()
ax.scatter(categories, percentages, color='green')
ax.set_ylabel("Percentage")
ax.set_title("Scatter Plot")
plt.xticks(rotation=45)
plt.show()

# 6. Stacked Bar Chart
fig, ax = plt.subplots()
cumulative = np.cumsum(percentages)
ax.bar(categories, cumulative, color=plt.cm.tab10.colors)
ax.set_ylabel("Cumulative Percentage")
ax.set_title("Stacked Bar Chart")
plt.xticks(rotation=45)
plt.show()

# 7. Area Chart
fig, ax = plt.subplots()
ax.fill_between(categories, percentages, color='skyblue', alpha=0.4)
ax.plot(categories, percentages, marker='o', color='blue')
ax.set_ylabel("Percentage")
ax.set_title("Area Chart")
plt.xticks(rotation=45)
plt.show()

# 8. Histogram
fig, ax = plt.subplots()
ax.hist(percentages, bins=5, color='purple', alpha=0.7, rwidth=0.85)
ax.set_xlabel("Percentage")
ax.set_ylabel("Frequency")
ax.set_title("Histogram")
plt.show()

# 9. Radar Chart
from math import pi

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
percentages += percentages[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, percentages, color='orange', alpha=0.25)
ax.plot(angles, percentages, color='orange', linewidth=2)
ax.set_yticks([10, 20, 30])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
ax.set_title("Radar Chart")
plt.show()

# 10. Donut Chart
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(percentages[:-1], labels=categories[:-1], autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)
ax.set_title("Donut Chart")
plt.show()
```
